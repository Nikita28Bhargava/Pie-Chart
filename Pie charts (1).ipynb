```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple pie chart
list_1=np.array([10,20,30,40])
plt.pie(list_1)
plt.show
```

```python
# Pie chart with labels
y=np.array([35,15,25,25])
mylabels=["Apple","Mango","Grapes","Banana"]

plt.pie(y,labels=mylabels)
plt.show()
```

```python
#exploder pie

y=np.array([35,15,25,25])
mylabels=["Men","Women","Boys","Girls"]
myexploded=[0.2,0,0,0]

plt.pie(y,labels=mylabels,explode=myexploded)
plt.show()
```

```python
# Shadow
y=np.array([10,20,30,40])
plt.pie(y,shadow=True)
plt.show()
```

```python
# Pie chart with colours
x=np.array([23,56,77,32])
mycolours=["orange","pink","blue","green"]
plt.pie(x,colors=mycolours)
plt.show()
```

```python
# Pie charts with legends
y=np.array([10,20,30,40])
mylabels=["india","japan","china","srilanka"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

```python
#legend with title
y=np.array([35,15,25,25])

mylabels=["india","japan","china","srilanka"]

plt.pie(y,labels=mylabels)
plt.legend(title="Countries")
plt.show()
```

```python
#pie chart with autopercentage
sports=["Cricekt","Hockey","Football","Tennis","Basketball"]

number_of_pepople=[500,240,300,109,308]

plt.pie(number_of_pepople,labels=sports,autopct="%.1f")
plt.legend(title="People Vs Sports")
plt.show()
```
